```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
import torch



def showres(path):
    img_data_tensor=torch.load(path)
    img_data= img_data_tensor[4].numpy()
    img_data.resize(56,56)
    plt.imshow(img_data)
    return img_data
```

```python
plt.subplot(1, 3, 1)
a1=showres('gal_img/sc62_x_100.pt')
plt.subplot(1, 3, 2)  
a2=showres('gal_img/sc62_x_hat_100.pt') 
plt.subplot(1, 3, 3) 
plt.imshow(a1-a2)
```
